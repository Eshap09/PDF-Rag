```python
import pypdf

path = "data/100_Must_Leetcode_Problems_1744774505.pdf"
reader = pypdf.PdfReader(path)

print(f"Total pages: {len(reader.pages)}\n")

for i, page in enumerate(reader.pages):
    text = page.extract_text() or ""
    text = text.strip()
    print(f"Page {i+1:2d} | {len(text):4d} chars | {repr(text[:80])}")
```

```python
def is_image_based_pdf(path: str) -> bool:
    reader = pypdf.PdfReader(path)
    total  = sum(len(p.extract_text() or "") for p in reader.pages)
    avg    = total / len(reader.pages)
    print(f"Avg chars/page: {avg:.1f}")
    return avg < 150    # raised threshold — below 150 is too sparse to be useful

path = "data/100_Must_Leetcode_Problems_1744774505.pdf"
if is_image_based_pdf(path):
    print("Too sparse — actual content is in images, needs OCR")
else:
    print("Has enough text to be useful")
```

```python
from app.rag.chunker import load_and_chunk

chunks = load_and_chunk("data/100_Must_Leetcode_Problems_1744774505.pdf", "leetcode.pdf")

print(f"Total chunks: {len(chunks)}\n")

for ch in chunks:
    print(f"chunk {ch.metadata['chunk_index']:2d} | page {ch.metadata.get('page_number')} | {len(ch.page_content):4d} chars")
    print(f"  {repr(ch.page_content[:100])}")
    print()
```

```python
import chromadb

client = chromadb.HttpClient(host='localhost', port=8001)
col    = client.get_collection('rag_documents')

print(f"Total chunks in DB: {col.count()}\n")

data = col.peek(100)

for doc, meta in zip(data['documents'], data['metadatas']):
    print(f"source={meta.get('source')} | page={meta.get('page_number')} | chunk={meta.get('chunk_index')}")
    print(f"  {repr(doc[:100])}")
    print()
```

```python
from app.rag.vectorstore import similarity_search

questions = [
    "what are the array problems?",
    "what graph problems should I solve?",
    "what is dynamic programming?",
    "what linked list problems are listed?",
]

for q in questions:
    print(f"Q: {q}")
    results = similarity_search(q, k=2)
    for r in results:
        print(f"  page={r.metadata.get('page_number')} | {repr(r.page_content[:80])}")
    print()
```
